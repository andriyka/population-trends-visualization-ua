# First visualization
As our data have many dimensions it's hard to place them all into a single chart, so I decided to build a small chart for each death cause showing age trends for it. 
Using such visualization, we are giving the most detailed overview to the user. 

```python
import numpy as np
grouped = df.groupby('death_cause')

ncols=4
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(28, 12))

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
    grouped.get_group(key).plot(ax=ax, x= 'age_group', y='death_count', kind = 'area', label = 'Death Count 0-70+')
    ax.set_aspect('auto')
    ax.title.set_text(key[:55])
    ax.set_frame_on(False)
    ax.get_xaxis().get_label().set_visible(False)

axes.flatten()[-1].axis("off")
fig.suptitle('Age(0 - 70+) distribution by each death cause', fontweight='bold')
#fig.subplots_adjust(top=0.93)

plt.show()
```

```python
main_causes = df.groupby('death_cause',  as_index=False).sum().sort_values(by='death_count', ascending = 0)[:4]
main_causes
```

```python
the_rest = df[~df.death_cause.isin(main_causes['death_cause'])]
the_rest = the_rest.groupby('age_group',  as_index=False).sum()
the_rest['death_cause'] = 'other'
the_rest
```

```python
main_df = df[df.death_cause.isin(main_causes['death_cause'])]
reduced_df = pd.concat([the_rest, main_df])
reduced_df.pivot_table(index='death_cause', columns='age_group', values='death_count')
```

```python
from colour import Color
gray = Color("black")
colors = list(gray.range_to(Color("white"),20))
colors = [c.hex for c in colors]
colors = colors[1:]
colors
```

# Second visualization
As the first one was in full details, I've decided to do the second one only with main death causes, so I've created the category - other and placed there all the death causes except five main. 
Afterward, I've built the distribution for each death cause by age and number of incidents. As it's three-dimensional data, I've chosen a stacked horizontal bar, where a length is a number of incidents and the color represents the age group. 

```python
import matplotlib.patches as mpatches

bars = []
patches = []
names = reduced_df['death_cause'].unique()
age_groups = reduced_df['age_group'].unique()

for ag in reduced_df.age_group.unique():
    bars.append(list(reduced_df[reduced_df.age_group == ag]['death_count']))
    
r = range(len(bars[0]))

plt.figure(figsize=(18,11))
plt.barh(r, bars[0], edgecolor='white', color = colors[0])
patches.append(mpatches.Patch(color=colors[0], label=age_groups[0]))


for i, bar in enumerate(bars[1:]):
    plt.barh(r, bar, edgecolor='white', left = list(map(sum, zip(*bars[:i+1]))), color = colors[i + 1])
    patches.append(mpatches.Patch(color=colors[i + 1], label=age_groups[i + 1]))

plt.yticks(r, names)
plt.xlabel("number of ")
plt.title("A distribution of deaths by age and cause. The strongest the color is - the oldest group it represents", fontweight='bold')
# Show graphic




red_patch = mpatches.Patch(color='red', label='The red data')
plt.legend(handles=patches)

plt.show()

plt.show()

```

# Third visualization
As we have seen on the previous chart, there is one main death cause and four others are much less present, so I've decided to split the data into two graphs and show trends for the main one and compare four others by age groups 

```python
grouped_reduced = reduced_df.groupby('death_cause')
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

grouped_reduced.get_group(' Хвороби системи кровообігу').plot(ax = ax1, x= 'age_group', y='death_count', label = ' Хвороби системи кровообігу')
plt.set_cmap(plt.cm.coolwarm)
plt.figure(figsize=(12,4))
for key in grouped_reduced.groups.keys():
    if key != ' Хвороби системи кровообігу':
        grouped_reduced.get_group(key).plot(ax = ax2, x= 'age_group', y='death_count', label = key)

plt.show()
```
