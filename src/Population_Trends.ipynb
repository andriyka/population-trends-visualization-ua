# Regions Comparison
As far as I'm concerned, the best way to show the comparison is a map. On the following map by the size of a circle described the value of birth/death rate. If the ring is red - population rate is negative otherwise it's positive. From this map, you can easily observe the full situation. In case you need precise info - there is a popup for each marker

```python
import folium

m = folium.Map([48, 31], zoom_start=6.49)
folium.GeoJson(geo_json_data).add_to(m)


for i in range(0,len(full_df_2016)):
   folium.Circle(
      location=[full_df_2016.lat[i], full_df_2016.lon[i]],
      popup=full_df_2016.index.values[i] + ' : ' + str(full_df_2016.rate[i]),
      radius=abs(full_df_2016.rate[i]*3000),
      color=('crimson' if full_df_2016.rate[i] < 0 else 'green'),
      fill=True,
      fill_color=('crimson' if full_df_2016.rate[i] < 0 else 'green')
   ).add_to(m)

m.save(os.path.join('..', 'UA_2016.html'))
m
```

# Regional Trends.
For each region, a small area chart is created. Using those chars - we could compare trends in regions and observe the overall situation.
This visualization is suitable as it provides detailed overview for each region as well as the overall one. 

```python
import numpy as np
grouped = df.groupby('region')

ncols=3
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(24, 18))

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
#     grouped.get_group(key).plot(ax=ax, x= 'year', y='rate', kind = 'area', label = key, stacked=False, color=list(grouped.get_group(key).rate.map(lambda v: 'r' if v <= 0 else 'g')))
    grouped.get_group(key).plot(ax=ax, x= 'year', y='rate',kind = 'area', label = key, stacked=False)
    ax.set_frame_on(False)
    ax.get_xaxis().get_label().set_visible(False)

axes.flatten()[-1].axis("off")
fig.suptitle('Динаміка приросту/скорочення населення за регіонами України', fontweight='bold')
fig.subplots_adjust(top=0.93)

plt.show()
```

# Comparison with Lviv
For this purpose, I've used the previous visualization and added a trend for Lviv region to each graph. To my mind, it's better than to do one graph with all trends as it will be really noisy.

```python
import numpy as np
grouped = df.groupby('region')

ncols=3
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(24, 18))

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
#     grouped.get_group(key).plot(ax=ax, x= 'year', y='rate', kind = 'area', label = key, stacked=False, color=list(grouped.get_group(key).rate.map(lambda v: 'r' if v <= 0 else 'g')))
    grouped.get_group('Львівська область').plot(ax=ax, x= 'year', y='rate', label = 'Львівська область', stacked=False, color='red', linestyle='--')
    grouped.get_group(key).plot(ax=ax, x= 'year', y='rate' , label = key,  kind='area', stacked=False, color = 'gray')
    ax.set_frame_on(False)
    
    ax.get_xaxis().get_label().set_visible(False)

axes.flatten()[-1].axis("off")
fig.suptitle('Динаміка приросту/скорочення населення за регіонами України у порівнянні з Львівською областю', fontweight='bold')
fig.subplots_adjust(top=0.93)

plt.show()
```
